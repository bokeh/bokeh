```python
# Create our data, has a high degree of variability
x = np.arange(1000)
base = np.arange(0, 500)
vardata = (np.random.rand(1000)*10) * np.concatenate((base, base[::-1]))
f = figure(title="Variable Data")
f.circle(x, vardata)
show(f)
```

```python
def get_bins(x, y, bins):
    '''Put data into bins'''
    prev = 0
    for i in range(1, len(y) - 1, len(y) // bins):
        yield x[prev:i], y[prev:i]
        prev = i
    yield x[prev:], y[prev:]
```

```python
def linear_fit(x, y, bins=None):
    '''Return the linear variability in a number of bins'''
    if bins is None:
        A = np.vstack([x, np.ones(len(x))]).T
        m, c = np.linalg.lstsq(A, y)[0]
        return m * x + c
    else:
        data = []
        for bx, by in get_bins(x, y, bins):
            data.append(linear_fit(bx, by))
        return np.concatenate(data)

# We want to measure "linear variability" -- or the variability from a linear match
# in a certain number of "bins"
varlines = linear_fit(x, vardata, 25)
f.line(x, varlines, line_color='red')
show(f)
```

```python
# Now we want to know the variability
# We will just subtract the line from the data and get the normalizesd variability
def linear_std(x, y, bins, line=None):
    '''Returns the normalized standard deviation of the data matched to *bins* lines'''
    # get unit mean
    unit_mean = np.mean(y - np.min(y))
    # normalize it to be around the center
    if line is None:
        line = linear_fit(x, y, bins)
    data = line - y
    # return the normalized linear standard deviation
    return np.std(data) / unit_mean

print("some bins  ", linear_std(x, vardata, 5))
print("many bins  ", linear_std(x, vardata, 30))
print("tons bins  ", linear_std(x, vardata, 200))

print("linear data", linear_std(x, np.arange(len(x)), 30))
```

```python
# Now, this data is all over the place, so we want to plot it in a bar and whisker plot
def whisker(fig, x, y, bins=None, line_width=1, color='blue', name=None):
    means = []
    for bx, by in get_bins(x, y, bins):
        # find the quantiles
        vlower, low, mid, up, vupper = np.percentile(by, [10, 25, 50, 75, 90])
        xpoint = np.mean(bx)
        # plot the whisker
        s = fig.segment(xpoint, vlower, xpoint, vupper, line_width=line_width, line_color=color)
        # plot the box
        xlen = max(bx) - min(bx)
        fig.rect(xpoint, np.mean([low, up]), xlen/10, up - low, fill_color=color, line_color=color)
        
        # plot the outliers
        outliers = np.concatenate((by[by < vlower], by[by > vupper]))
        fig.circle(np.repeat(xpoint, len(outliers)), outliers, radius=line_width, fill_color=color, line_color=color)
        
        # add the mean to plot as a line later
        means.append((xpoint, mid))
    
    fig.line(*zip(*means), line_color=color, legend=name)


f = figure(title="Auto Box-Whisker")
# f.circle(x, vardata)
whisker(f, x, vardata, 15, color='red', name='test')
show(f)
```

```python
# The hope is that this will end up in the charts interface for bokeh
def dataframe(fig, df,
              variability_cutoff: "data with variabilty > this will be box-whisker"=0.15,
              min_bins: "The minimum number of bins"=10, 
              bins: "the desired number of bins"=30, 
              num_per_bin: "dataframe will try to have at least this many data points per bin"=20,
              min_num_per_bin: "any number less than this will just be plotted as a scatter plot"=5,
              color='blue'):
    
    for color, col in zip(get_colors(len(df.columns)), df.columns):
        scat = False
        data = df[col].dropna()
        if len(data) / min_bins <= num_per_bin:
            _bins = min_bins
        else:
            _bins = bins
        if len(data) / min_bins < min_num_per_bin:
            scat = True
        if not scat:
            line = linear_fit(data.index, data, _bins)
            std = linear_std(data.index, data, _bins, line=line)
            if std <= variability_cutoff:
                scat = True
                
        if scat:
            scatter(fig, data.index, data, _bins, color=color, name=col)
        else:
            whisker(fig, data.index, data, _bins, color=color, name=col)
            
f = figure()
dataframe(f, df)
show(f)
```
